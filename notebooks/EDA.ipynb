```python
import pandas as pd
import plotly.express as px

# Carregar o arquivo CSV
df = pd.read_csv('../dados/enem_amostra.csv')

# Verificar primeiras linhas
df.head()

# Verificar dados ausentes
df.isna().sum()

# Verificar tipos de dados
df.info()

# Filtro: alunos presentes nos dois dias
df_validos = df[(df['TP_PRESENCA_D1'] == 1) & (df['TP_PRESENCA_D2'] == 1)]

# Histograma da idade
px.histogram(df, x='NU_IDADE', nbins=20, title='Distribuio das Idades').show()

# Boxplot por sexo
px.box(df_validos, x='TP_SEXO', y='NU_NOTA_MT', title='Nota de Matemtica por Sexo').show()

# Relao entre Humanas e Exatas
px.scatter(df_validos, x='NU_NOTA_CH', y='NU_NOTA_CN', color='TP_SEXO',
           title='Nota de Humanas vs Cincias da Natureza por Sexo').show()
```
